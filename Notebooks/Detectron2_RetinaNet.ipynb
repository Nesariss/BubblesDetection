```python
DATASET_FOLDER_NAME = "bubles222-1"
DETECTRON2_MODEL_NAME = "retinanet.pth"
```

## Import and Register Custom Detectron2 Data

```python
import detectron2
from detectron2.utils.logger import setup_logger

setup_logger()

import numpy as np
import cv2
import random
from google.colab.patches import cv2_imshow

from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
from detectron2.data.catalog import DatasetCatalog
```

```python
# To work with datasets from Detectron2 we need to intialize object with appropriate format
# or use the method register_coco_instances in case we already do have the dataset in coco json format
# Notice: needs to be run only once

from detectron2.data.datasets import register_coco_instances


TRAIN_DATASET_NAME = f"{DATASET_FOLDER_NAME}_dataset_train"
VALID_DATASET_NAME = f"{DATASET_FOLDER_NAME}_dataset_valid"
TEST_DATASET_NAME = f"{DATASET_FOLDER_NAME}_dataset_test"

register_coco_instances(TRAIN_DATASET_NAME, {}, f"{DATASET_FOLDER_NAME}/train/_annotations.coco.json", f"{DATASET_FOLDER_NAME}/train")
register_coco_instances(VALID_DATASET_NAME, {}, f"{DATASET_FOLDER_NAME}/valid/_annotations.coco.json", f"{DATASET_FOLDER_NAME}/valid")
register_coco_instances(TEST_DATASET_NAME, {}, f"{DATASET_FOLDER_NAME}/test/_annotations.coco.json", f"{DATASET_FOLDER_NAME}/test")
```

```python
# Visualize the registered training data

import random
from detectron2.utils.visualizer import Visualizer


METADATA_TRAIN_DATASET = MetadataCatalog.get(TRAIN_DATASET_NAME)
DATASET_CATALOG = DatasetCatalog.get(TRAIN_DATASET_NAME)


for random_sample in random.sample(DATASET_CATALOG, 3):
    image = cv2.imread(random_sample["file_name"])
    visualizer = Visualizer(image[:, :, ::-1], metadata=METADATA_TRAIN_DATASET, scale=1)
    vis = visualizer.draw_dataset_dict(random_sample)
    cv2_imshow(vis.get_image()[:, :, ::-1])
```

## Train Custom Detectron2 Detector

```python
from detectron2.engine import DefaultTrainer
from detectron2.evaluation import COCOEvaluator


class CocoTrainer(DefaultTrainer):

  @classmethod
  def build_evaluator(cls, cfg, dataset_name, output_folder=None):

    if output_folder is None:
        os.makedirs("coco_eval", exist_ok=True)
        output_folder = "coco_eval"

    return COCOEvaluator(dataset_name, cfg, False, output_folder)
```

```python
# Model initialization

from detectron2.config import get_cfg
import os



#CONFIG_FILE = "COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"
CONFIG_FILE ="COCO-Detection/retinanet_R_50_FPN_1x.yaml"

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file(CONFIG_FILE))
cfg.DATASETS.TRAIN = (TRAIN_DATASET_NAME,)
cfg.DATASETS.TEST = (VALID_DATASET_NAME,)
cfg.DATALOADER.NUM_WORKERS = 4

# In case of training from the beginning, the same path is stated as in initialization
# In case of loading model from this place, state the path to file with save weights

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(CONFIG_FILE)
cfg.SOLVER.IMS_PER_BATCH = 4
cfg.SOLVER.BASE_LR = 0.001

# Setting the number of epochs
cfg.SOLVER.WARMUP_ITERS = 100
cfg.SOLVER.MAX_ITER = 1500
cfg.SOLVER.STEPS = (1000, 1501)
cfg.SOLVER.GAMMA = 0.05
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64

# For LDANN dataset we have 2 classes
#cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
cfg.MODEL.RETINANET.NUM_CLASSES = 2

cfg.TEST.EVAL_PERIOD = 500


os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
trainer = CocoTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```
